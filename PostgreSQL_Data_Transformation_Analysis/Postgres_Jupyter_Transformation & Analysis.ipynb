##
üî∏ if_exists='append'

This controls what happens if the table already exists in the database.
Behavior
'fail'
‚ùå Throws an error if the table already exists.
'replace'
‚ö†Ô∏è Drops the table if it exists, recreates it, then inserts the data.
'append'
‚úÖ Adds the data to the existing table without deleting anything.
#
üî∏ index=False

This controls whether the DataFrame‚Äôs index column is written to the database as a separate column.
	‚Ä¢	If index=True ‚Üí a new column like index will be created in the table.
	‚Ä¢	If index=False ‚Üí the index is ignored, and only the DataFrame‚Äôs actual columns are used.

In your case, index=False is correct, since your passenger_id is already part of the data.
