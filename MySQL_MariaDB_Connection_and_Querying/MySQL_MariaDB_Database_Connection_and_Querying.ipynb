```python
%%sql
SELECT nationality,travel_class, COUNT(*) AS total_passengers
FROM aero_passengers
WHERE nationality IN (
    SELECT nationality
    FROM aero_passengers
    GROUP BY nationality
    HAVING COUNT(*) > 1
)
GROUP BY nationality order by total_passengers desc
```

```python
%%sql
SELECT count(distinct nationality), count(distinct travel_class), COUNT(*) AS total_passengers
FROM aero_passengers
WHERE nationality IN (
    SELECT nationality
    FROM aero_passengers
    GROUP BY nationality
    HAVING COUNT(*) > 1
)
GROUP BY nationality, travel_class order by total_passengers desc
```

```python
%%sql SELECT # Sql pivot like case statement written in sql using count and case statements with grouo by and order by clauses #
    nationality,
    COUNT(CASE WHEN travel_class = 'First' THEN 1 END) AS first_class_passengers,
    COUNT(CASE WHEN travel_class = 'Business' THEN 1 END) AS business_class_passengers,
    COUNT(CASE WHEN travel_class = 'Economy' THEN 1 END) AS economy_class_passengers,
    COUNT(*) AS total_passengers
FROM 
    aero_passengers
GROUP BY 
    nationality
ORDER BY 
    total_passengers DESC;
```
